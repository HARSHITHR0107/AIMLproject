```python
#importing required packages
import keras
from keras.layers import Input, Dense
from keras import regularizers
import tensorflow as tf
from keras.models import Model
from sklearn import metrics
```

```python
#building autoencoder model

input_dim = X_train.shape[1]
encoding_dim = input_dim

input_layer = Input(shape=(input_dim, ))
encoder = Dense(encoding_dim, activation="relu",
                activity_regularizer=regularizers.l1(10e-4))(input_layer)
encoder = Dense(int(encoding_dim), activation="relu")(encoder)

encoder = Dense(int(encoding_dim-2), activation="relu")(encoder)
code = Dense(int(encoding_dim-4), activation='relu')(encoder)
decoder = Dense(int(encoding_dim-2), activation='relu')(code)

decoder = Dense(int(encoding_dim), activation='relu')(encoder)
decoder = Dense(input_dim, activation='relu')(decoder)
autoencoder = Model(inputs=input_layer, outputs=decoder)
autoencoder.summary()
```

```python
#compiling the model
autoencoder.compile(optimizer='adam',
                    loss='binary_crossentropy',
                    metrics=['accuracy'])

#Training the model
history = autoencoder.fit(X_train, X_train, epochs=10, batch_size=64, shuffle=True, validation_split=0.2)
```

```python
acc_train_auto = autoencoder.evaluate(X_train, X_train)[1]
acc_test_auto = autoencoder.evaluate(X_test, X_test)[1]

print('\nAutoencoder: Accuracy on training Data: {:.3f}' .format(acc_train_auto))
print('Autoencoder: Accuracy on test Data: {:.3f}' .format(acc_test_auto))
```

```python
#storing the results. The below mentioned order of parameter passing is important.
#Caution: Execute only once to avoid duplications.
storeResults('AutoEncoder', acc_train_auto, acc_test_auto)
```

```python
from urllib.parse import urlparse
import matplotlib.pyplot as plt

def classify_url(url):
    try:
        parsed_url = urlparse(url)
        domain = parsed_url.netloc
        path = parsed_url.path

        if not domain:
            return "Invalid URL(Phishing Website)"

        have_ip = 1 if any(char.isdigit() for char in domain.split('.')) else 0
        have_at = 1 if "@" in url else 0
        url_length = 1 if len(url) >= 54 else 0
        url_depth = path.count('/')
        redirection = 1 if "//" in path else 0
        https_domain = 1 if url.startswith("https") else 0
        tiny_url = 1 if len(url) <= 20 else 0
        prefix_suffix = 1 if "-" in domain else 0
        dns_record = 1
        web_traffic = 1
        domain_age = 1
        domain_end = 1
        iframe = 0
        mouse_over = 0
        right_click = 1
        web_forwards = 0

        features = {
            "Have IP": have_ip,
            "Have @": have_at,
            "URL Length": url_length,
            "URL Depth": 1 if url_depth > 4 else 0,
            "Redirection": redirection,
            "HTTPS Domain": 1 - https_domain,
            "Tiny URL": tiny_url,
            "Prefix/Suffix": prefix_suffix,
            "DNS Record": 1 - dns_record,
            "Web Traffic": 1 - web_traffic,
            "Domain Age": 1 - domain_age,
            "Domain End": 1 - domain_end,
            "iFrame": iframe,
            "Mouse Over": mouse_over,
            "Right Click": 1 - right_click,
            "Web Forwards": web_forwards,
        }

        phishing_score = sum(features.values())

        # Classification
        classification = "Phishing Website" if phishing_score > 5 else "Legitimate Website"

        # Visualization
        plt.figure(figsize=(10, 6))
        plt.bar(features.keys(), features.values(), color='blue')
        plt.xticks(rotation=45, ha='right')
        plt.title(f"Feature Contributions to Classification ({classification})")
        plt.xlabel("Features")
        plt.ylabel("Values")
        plt.tight_layout()
        plt.show()

        return classification

    except Exception as e:
        return f"Error processing URL: {e}"


# Example usage
url = input("Enter a URL: ")
result = classify_url(url)
print(f"The URL is classified as: {result}")
```
